```python
# função que forma o eixo y em valores de quantidade a partir dos df dos meses especificados; busacando a descricao dita

def formar_eixoY(dataFrames: list, nomeCaneta: str) -> list:
    eixoY = []
    for df in dataFrames:
        eixoY.append(df[df["Descricao"] == nomeCaneta].values[0][1])
    
    return eixoY

# eixo X de todas as analises abaixo
eixoX = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio", "Junho", "Julho"]
```

VENDAS DA CANETA BIC CRISTAL AZ EM CADA MÊS

```python
eixoY = formar_eixoY([csv_jan, csv_feb, csv_mar, csv_apr, csv_may, csv_jun, csv_jul], "CANETA BIC CRISTAL AZ")

plt.figure(figsize=(10, 6))
plt.bar(eixoX, eixoY)

for barra in plt.bar(eixoX, eixoY):
    plt.text(barra.get_x() + barra.get_width() / 2, barra.get_height(), f"{int(barra.get_height())}", ha = "center", va="bottom")

plt.xlabel("Meses")
plt.ylabel("Quantidade de vendas")
plt.title("Vendas da CANETA BIC CRISTAL AZ por mês")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

print(f"Mediana de vendas por mês: {int(np.median(eixoY))}")
print(f"Média de vendas por mês: {round(np.mean(eixoY), 2)}")
print(f"Desvio padrão das vendas por mês: {round(np.std(eixoY, ddof=1), 2)}")
```
